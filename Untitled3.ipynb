```python
df = load_wine(as_frame=True)
```

```python
df.data.plot(kind='kde', subplots=True, layout=(4, 4), figsize=(14, 14), sharey=False, sharex=False, legend=False, title=df.data.columns.tolist());
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(df.data, df.target, random_state=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier().fit(Xtrain, ytrain)
```

```python
clf.score(Xtest, ytest)
```
